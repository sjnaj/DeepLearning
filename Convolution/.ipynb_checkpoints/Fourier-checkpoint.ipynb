```matlab

E=1;a=1;
t=0:0.01:4;
w=-30:0.01:30;
f=E*-exp(-a*t);
F=1./(a+j*w);
plot(t,f);
xlabel('t');
ylabel('f(t)');
figure
plot(w,abs(F));
xlabel('\omega');
ylabel('|F(\omega)|');
figure;
plot(w,20*log10(abs(F)));
xlabel('\omega');
ylabel('|F(\omega)| in dB');
figure;
plot(w,angle(F));
xlabel('\omega');
ylabel('\phi \omega')
```
